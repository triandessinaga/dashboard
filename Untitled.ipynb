```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import streamlit as st
```

```python
# Load dataset
df = pd.read_csv("PRSA_Data_Aotizhongxin_20130301-20170228.csv")  # Gantilah dengan nama file kamu
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
```

```python
# Cek informasi dataset
df.info()

# Menampilkan beberapa baris data untuk melihat struktur dataset
df.head()
```

```python
# Memeriksa missing values
df.isnull().sum()

# Menghapus baris dengan nilai yang hilang (jika ada)
df.dropna(inplace=True)
```

```python
# Menampilkan tren PM2.5
sns.lineplot(data=df, x="datetime", y="PM2.5", hue="station")
plt.title("Tren PM2.5 per Jam")
plt.xlabel("Waktu")
plt.ylabel("PM2.5")
plt.show()
```

```python
# Time Series Plot
fig, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(data=df, x="datetime", y="PM2.5", hue="station", ax=ax, palette="flare")
ax.set_title("Hourly PM2.5 Levels")
ax.set_ylabel("PM2.5")
ax.set_xlabel("Time")
plt.show()
```

```python
# CO vs NO2 Correlation
fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=df, x="CO", y="NO2", hue="station", style="station", palette="rocket", ax=ax)
ax.set_title("CO vs NO2 Concentrations")
ax.set_xlabel("CO")
ax.set_ylabel("NO2")
plt.show()
```

```python
# Wind Speed Distribution
fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x="station", y="WSPM", palette="coolwarm", ax=ax)
ax.set_title("Wind Speed (WSPM) per Station")
plt.show()
```
